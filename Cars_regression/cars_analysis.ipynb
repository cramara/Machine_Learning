###  Import des données dans un tableau pandas

```python
import csv
import pandas as pd

data = pd.read_csv("./files/train.csv")

data.head()

```

Nettoyage des données : 

```python
print(f"Initial: {len(data)}")
# Affiche les informations générales sur le DataFrame (types, valeurs non nulles, etc.)
data.info()

# Calcule les statistiques descriptives pour les colonnes numériques
data.describe()

# Compte le nombre de valeurs nulles dans chaque colonne
data.isnull().sum()

# Affiche le nombre de valeurs nulles par colonne pour mieux comprendre
print("Nombre de valeurs nulles par colonne:")
print(data.isnull().sum())

# On peut choisir de ne supprimer que les lignes où certaines colonnes importantes sont nulles
colonnes_importantes = ['id', 'brand', 'model', 'car_class', 'fuel_type','range',
'max_power', 'overall_cons', 'co2','grbx_type_ratios', 'weight_min', 'weight_max',
'urb_cons', 'exturb_cons']
data.dropna(subset=colonnes_importantes, inplace=True)
print(f"After removing null in important columns: {len(data)}")

# Ou remplacer les valeurs nulles par des valeurs appropriées
data['hc'].fillna(data['hc'].mean(), inplace=True)
data['hcnox'].fillna(data['hcnox'].mean(), inplace=True)
data['ptcl'].fillna(data['ptcl'].mean(), inplace=True)

# On pourrait supprimer les lignes ou nox et co sont nulles (car seulement 86 lignes)
data['nox'].fillna(data['nox'].mean(), inplace=True)
data['co'].fillna(data['co'].mean(), inplace=True)


print(f"After handling nulls: {len(data)}")
print(data.isnull().sum())

# Compte le nombre de lignes dupliquées
data.duplicated().sum()

# Supprime les lignes dupliquées et modifie le DataFrame original
data.drop_duplicates(inplace=True)
print(f"After removing duplicates: {len(data)}")
```

```python
data.to_csv('./files/data_cleaned_train.csv')
```

Le nettoyage des données est effectué, on peut alors sauvegarder les données propres dans un csv
